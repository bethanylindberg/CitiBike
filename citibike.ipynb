Weekdays comparison:

We can assume that tourists’  usage of Citibike stays consistent during the week. So we can calculate average number of rides for Saturday and Sunday to get Recreational usage. We can also calculate average number of rides Monday-Friday. The difference between 2 would be roughly equal to usage for  Business purposes.

```python
df = pd.read_csv('tripdata.csv')
# Add column for weekdays
df["Weekday"] = df["starttime"].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f') if pd.notnull(x) else np.nan)\
                                .map(lambda x: calendar.day_name[x.weekday()]  if pd.notnull(x) else np.nan)
df
```

```python
# Calculate Number of rides by Weekend and Weekday
weekend_days = df.loc[df['Weekday'].isin(['Saturday','Sunday'])]['date'].nunique()
weekdays = df.loc[~df['Weekday'].isin(['Saturday','Sunday'])]['date'].nunique()

weekend_rides = df.loc[df['Weekday'].isin(['Saturday','Sunday'])]['Weekday'].count()
weekday_rides = df.loc[~df['Weekday'].isin(['Saturday','Sunday'])]['Weekday'].count()
```

```python
weekday_daily_average = round(weekday_rides/weekdays,0)
weekend_daily_average = round(weekend_rides/weekend_days,0)

approximate_daily_business = weekday_daily_average - weekend_daily_average

ride_summary = pd.DataFrame({"Avg Daily Sat-Sun":int(weekend_daily_average),\
                         "Avg Weekday":int(weekday_daily_average),\
                         "Approx. Daily Business Rides":int(approximate_daily_business)},index=[""])
ride_summary
```
